### 6.)Date Conversion and Lead Time Calculation & Calculating key statistics for Shipping Lead Time.
 
Converting Order Date and Ship Date from object (text) to datetime format.
This allows us to perform date arithmetic to calculate Shipping Lead Time.

Formula: Lead Time = Ship Date - Order Date

```python
# Converting date columns from object to datetime format
df['Order Date']=pd.to_datetime(df['Order Date'],format='%d-%m-%Y')
df['Ship Date']=pd.to_datetime(df['Ship Date'],format='%d-%m-%Y')

# Calculating shipping lead time in days
# Lead Time = number of days between order placement and shipment
df['Lead Time']=abs(df['Ship Date']-df['Order Date']).dt.days

# Displaying lead time statistics
print(f"Minimum Lead Time:{df['Lead Time'].min()} days")
print(f"Maximum Lead Time:{df['Lead Time'].max()} days")
print(f"Average Lead Time:{df['Lead Time'].mean():.2f} days")
print(f"Most Common Lead Time:{df['Lead Time'].mode()[0]} days") #Unimodal(Lead time)
```

### 6.)Finding:
Lead Time column successfully created for all orders,Values represent the number of days between order placement and shipment for each record.
The Minimum Lead Time Observed:904 days
The Maximum Lead Time Observed :1642 days
The average shipping Lead time across all orders is 1321 days.
The maximum Lead time observed is 1642 days which indicates potential delays on certain routes.

### 9.)Extracting Product to Factory Mapping, Mapping Products to Factories

Every product in the Nassau Candy dataset is manufactured
at a specific factory. Since the dataset does not contain
a Factory column directly, we load the product-factory
correlation from an external CSV file and map it to our
main dataset using Product Name as the key.

This step is critical because our entire route analysis
depends on knowing which factory each order originated from.

Mapping Key: Product Name → Factory Name
I will extract the unique product and division Combinations directly from the dataset itself. 

```python
# first lets see what unique products and divisions exist
# in our dataset
print(df[['Product Name', 'Division']].drop_duplicates().sort_values('Division'))
```

```python
### Loading Factory Dataset
# factory mapping data is sourced from the project description
# stored as an external csv file to keep data and code separate
factory_df = pd.read_csv(r"A:\UM\nassau_candy_project/data/factory_mapping.csv")

# converting the csv into a dictionary
# key = Product Name, value = Factory Name
factory_dic = dict(zip(factory_df['Product Name'], factory_df['Factory']))

# applying the dictionary mapping to create Factory column
# every product name gets replaced with its factory name
df['Factory'] = df['Product Name'].map(factory_dic)

# verifying mapping worked correctly
print(f"Unmapped Products: {df['Factory'].isnull().sum()}")
print(df[['Product Name','Factory']].drop_duplicates())
```

### 9.) Finding
All 10194  products successfully mapped to their respective factories.
Zero unmapped products confirms complete and accurate factory assignment.

- Lot's O' Nuts

## # Step 11 - Creating Route Column

A route is defined as the shipping path from a factory
to a customer state. This is the most important feature
of our entire analysis. Every efficiency insight, KPI
and bottleneck analysis will be based on this column.

Formula: Route = Factory Name → Customer State

```python
# combining factory name and customer state
# to create a unique route identifier for each order
df['Route'] = df['Factory'] + ' → ' + df['State/Province']

# also creating a region level route
# for higher level geographic analysis
df['Region Route'] = df['Factory'] + ' → ' + df['Region']

# checking how many unique routes exist
print(f"Total Unique State Routes: {df['Route'].nunique()}")
print(f"Total Unique Region Routes: {df['Region Route'].nunique()}")

# checking top 10 most common routes
print("\nTop 10 Most Common Routes:")
print(df['Route'].value_counts().head(10)) #Since Value_counts() arranges in desecnding order
```
